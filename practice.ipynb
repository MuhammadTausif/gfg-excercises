```python
import PyPDF2
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def extract_text_from_pdf(input_pdf_path):
    """Extract all text from the PDF."""
    text = ""
    with open(input_pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text

def save_text_to_pdf(text, output_pdf_path):
    """Save the extracted text to a new PDF file using ReportLab."""
    c = canvas.Canvas(output_pdf_path, pagesize=letter)
    width, height = letter

    # Start a text object at a margin (40 points from the left and top)
    text_obj = c.beginText(40, height - 40)
    text_obj.setFont("Helvetica", 10)

    # Process the text line by line
    for line in text.splitlines():
        text_obj.textLine(line)
        # If we are too close to the bottom, finish the page and start a new one.
        if text_obj.getY() < 40:
            c.drawText(text_obj)
            c.showPage()
            text_obj = c.beginText(40, height - 40)
            text_obj.setFont("Helvetica", 10)

    # Draw the final text and save the PDF
    c.drawText(text_obj)
    c.save()
```

```python
import os
import re
from collections import defaultdict

def tokenize(text):
    """
    Tokenizes the text into alphanumeric words.
    Returns a list of tokens in lowercase.
    """
    return re.findall(r'\b\w+\b', text.lower())

def build_inverted_index(folder, num_files=5):
    """
    Reads up to num_files .txt files from the given folder.
    Builds an inverted index mapping terms (unigrams and bigrams)
    to a list of document ids where the term appears.
    """
    inverted_index = defaultdict(set)
    
    # List only .txt files in the folder and limit to num_files.
    files = sorted([f for f in os.listdir(folder)
                    if f.lower().endswith(".txt") and os.path.isfile(os.path.join(folder, f))])[:num_files]
    
    for doc_id, filename in enumerate(files, 1):
        file_path = os.path.join(folder, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            tokens = tokenize(content)
            
            # Add unigrams to the inverted index.
            for token in tokens:
                inverted_index[token].add(doc_id)
            
            # Add bigrams (two consecutive words) to the inverted index.
            for i in range(len(tokens) - 1):
                bigram = tokens[i] + " " + tokens[i+1]
                inverted_index[bigram].add(doc_id)
    
    # Convert each set of document ids to a sorted list.
    return {term: sorted(list(doc_ids)) for term, doc_ids in inverted_index.items()}

# Specify the folder path containing your .txt files.
folder_path = r"C:\Users\Public\Dev\Ph.D\2nd Semester\CS-675-IRS\Assignments\First\IRS (101-105)\Text"  # Replace with your folder path

# Build the inverted index from five .txt files.
inverted_index = build_inverted_index(folder_path, num_files=5)

# Print the inverted index in a readable format.
print("{")
for term, doc_ids in inverted_index.items():
    print(f'    "{term}": {doc_ids},')
print("}")
```

```python
import os
import PyPDF2

def extract_text_from_pdf(input_pdf_path):
    """Extract all text from the PDF."""
    text = ""
    with open(input_pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text

def save_text_to_file(text, output_txt_path):
    """Save the extracted text to a text file."""
    with open(output_txt_path, "w", encoding="utf-8") as file:
        file.write(text)

# Specify the folder path containing your PDF files.
pdf_folder_path = r"C:\Users\Public\Dev\Ph.D\2nd Semester\CS-675-IRS\Assignments\First\IRS (101-105)\PDF"  # Replace with your folder path
pdf_folder_path = r"C:\Users\Public\Dev\Ph.D\2nd Semester\CS-675-IRS\Assignments\First\PDF"
# Loop over each PDF file in the folder and convert to text file.
for file in os.listdir(pdf_folder_path):
    if file.lower().endswith(".pdf"):
        input_pdf_path = os.path.join(pdf_folder_path, file)
        output_txt_path = os.path.join(pdf_folder_path, file.replace(".pdf", ".txt"))
        
        # Extract text from the PDF and save to a text file.
        extracted_text = extract_text_from_pdf(input_pdf_path)
        save_text_to_file(extracted_text, output_txt_path)

print("PDF files have been converted to text files.")
```
